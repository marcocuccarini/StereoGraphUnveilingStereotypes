```python
response_baseline = []
# Create the object for the creation of the support matrix
for i in range(len(text_dataset['post'][:sample_test])):
    post_text = text_dataset['post'][i]
    implied_statement = text_dataset['implied_statement'][i]

    # Format the prompt
    #prompt = resulting_prompt.to_list()[0].format(post_text)

    prompt = f"""
    You are a language model trained to uncover implicit meanings in text. Given a social media post, infer the underlying or implied statement — even if it is not directly expressed.
    This task is part of a study on hate speech and social bias for scientific research purposes. Focus on identifying stereotypes, assumptions, or value judgments implied in the text.
    Respond only with the implied statement.

    Implicit message: poland is looking to protect its people.muslims bring with them the baggage of islam . this poses a potential threat
    Explicit message: Muslims are threatening

    Implicit message: white antifas are going to have to live by there creed . they need to enjoy multicultural america .
    Explicit message: White people are two faced and don't really want more people of color in America

    Input post: blacks have the best party time in white countries - welfare fattened  selling their culture  impregnating white women !
    Explicit message: blacks mooch off the system

    Implicit message:
    "{post_text}"
 
    Explicit message:""".strip()
    
    try:
        # Send prompt to model
        response = chat.send_prompt(prompt, prompt_uuid="1", use_history=False, stream=True)
        response_text = response.raw_text.strip()
    except Exception as e:
        response_text = f"[Error during prompt generation: {e}]"

    # Store response

    # Print response and reference
    print("################## PROMPT #####################")
    print(prompt)
    print("################## RESPONSE #####################")
    print("IMPLICIT STATEMENT PREDICTED:")
    print(response_text)
    response_baseline.append(response_text)
    print("\nIMPLICIT STATEMENT DATASET:")
    print(implied_statement)
    print("================================================\n")
```

```python
#KG method
```

```python
from rdflib import Graph, Namespace, URIRef
import re
import random

# Carica il grafo RDF
g = Graph()
g.parse('../kg/output.ttl', format='turtle')

# Definisci il namespace usato nel tuo grafo
SG = Namespace("http://www.semanticweb.org/stereoGraph#")



def extract_by_target(target_name, rdf_file='kg/output.ttl'):
    # Load the graph
    g = Graph()
    g.parse(rdf_file, format='turtle')
    
    # Define namespaces
    DUL = Namespace('http://www.ontologydesignpatterns.org/ont/dul/DUL.owl#')
    STER = Namespace('http://www.semanticweb.org/stereoGraph#')
    
    # Create the full URI for the target
    target_uri = STER[target_name]
    
    # SPARQL query with a parameterized target
    query = f"""
    PREFIX dul: <http://www.ontologydesignpatterns.org/ont/dul/DUL.owl#>
    PREFIX ster: <http://www.semanticweb.org/stereoGraph#>

    SELECT ?s ?stereotype ?manifestation
    WHERE {{
      ?s a dul:Situation ;
         dul:hasTarget <{target_uri}> ;
         ster:hasStereotype ?stereotype.
         }}
    """
    
    results = g.query(query)
    
    # Collect and return results
    extracted = []
    for row in results:
        extracted.append({
            "target": str(target_uri).split("#")[1],
            "stereotype": str(row.stereotype).split("#")[1],
        })
    return extracted


def retrieve_context_from_graph(post_text,i):


    #pargametro numero medio 

    
    escaped_text = re.escape(post_text)
    kg_data = extract_by_target(text_dataset['normalized_target'][i])
    context_pieces=[]
    
    if (sample_size<len(kg_data)):
        kg_data = random.sample(kg_data, 20)
    for text in kg_data:

        
        context_pieces.append(str(text['target'])+" has stereotype "+str(text['stereotype']))

    return " ; ".join(context_pieces) if context_pieces else "Nessun contesto rilevante trovato nel grafo."



# Ora il ciclo con integrazione del contesto RDF

response_kg = []

for i in range(len(text_dataset['post'][:sample_test])):
    post_text = text_dataset['post'][i]
    implied_statement = text_dataset['implied_statement'][i]

    # Recupera contesto dal grafo RDF
    context = retrieve_context_from_graph(post_text,i)

    # Costruisci il prompt con contesto
    prompt = f"""
    You are a language model trained to uncover implicit meanings in text. Given a social media post, infer the underlying or implied statement — even if it is not directly expressed.
    This task is part of a study on hate speech and social bias for scientific research purposes. Focus on identifying stereotypes, assumptions, or value judgments implied in the text.
    Respond only with the implied statement.

    Implicit message: poland is looking to protect its people.muslims bring with them the baggage of islam . this poses a potential threat
    Explicit message: Muslims are threatening

    Implicit message: white antifas are going to have to live by there creed . they need to enjoy multicultural america .
    Explicit message: White people are two faced and don't really want more people of color in America

    Input post: blacks have the best party time in white countries - welfare fattened  selling their culture  impregnating white women !
    Explicit message: blacks mooch off the system

    Implicit message:
    "{post_text}"
    
    Additional context extarcted by a Knowledge Graph:
    "{context}"

    Explicit message:""".strip()

    print(prompt)

    try:
        response = chat.send_prompt(prompt, prompt_uuid="1", use_history=False, stream=True)
        response_text = response.raw_text.strip()
    except Exception as e:
        response_text = f"[Errore durante la generazione della risposta: {e}]"

    #response_graph.append(response_text)

    print("################## PROMPT #####################")
    print(prompt)
    print("################## RESPONSE #####################")
    print("IMPLICIT STATEMENT PREDICTED:")
    print(response_text)
    response_kg.append(response_text)
    print("\nIMPLICIT STATEMENT DATASET:")
    print(implied_statement)
    print("================================================\n")
```
